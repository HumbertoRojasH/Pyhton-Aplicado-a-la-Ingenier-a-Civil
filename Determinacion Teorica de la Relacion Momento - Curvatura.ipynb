# Concreto no Confinado - Modelo de Hognestad

Este modelo está compuesto de una parte curvilínea descrita por una parábola que comienza en cero y termina cuando el concreto a compresión alcanza la resistencia máxima f'c , en este punto la parábola es tangente a la horizontal. A partir de este punto los esfuerzos del concreto disminuyen linealmente conforme aumentan las deformaciones hasta que ocurre la falla. La expresión que representa este modelo es:

<div align="center"><img src="img/Modelo-no-confinado.svg"
     width="500"></div>

Donde $f_c$ y $\varepsilon_c$ es el esfuerzo y la deformación unitaria en el concreto, respectivamente, mientras que $\varepsilon _{co}$ es la deformación unitaria correspondiente a $f'_c$ y $\varepsilon_{cm}$ es la deformación unitaria máxima útil del concreto en compresión.

$$f_c=\left\{\begin{matrix} f'_c\left[ \frac{2\varepsilon }{\varepsilon_{co}}-\left ( \frac{\varepsilon }{\varepsilon_{co}} \right )^{2} \right ] & \hspace{1em}, 0\leqslant \varepsilon \leqslant \varepsilon_{co} \\ 
 f'c-0.15f'c\frac{(\varepsilon -\varepsilon _{co})}{(\varepsilon_{cm}-\varepsilon _{co})} & \hspace{2em}, \varepsilon _{co}< \varepsilon \leqslant \varepsilon_{cm}
\end{matrix}\right.$$

El coeficiente $\alpha$ representa un procentaje del área bajo la curva con respecto al área del rectángulo cuya área es $f'_c\varepsilon_c$.

$$\alpha =\frac{\int_{0}^{\varepsilon_c}f_cd\varepsilon}{f'_c\varepsilon_c}$$


$$\alpha =\left\{\begin{matrix}
 \frac{\varepsilon _c(3\varepsilon_{co}-\varepsilon_c)}{3{\varepsilon_{co}}^{2}} &, 0\leqslant \varepsilon \leqslant \varepsilon_{co} 
\\ 
 \frac{40\varepsilon_{co}\varepsilon_{cm}-31{\varepsilon_{co}}^{2}-120\varepsilon_{c}\varepsilon_{cm}+102\varepsilon_{c}\varepsilon_{co}+9{\varepsilon_{c}}^{2}}{120\varepsilon_c(\varepsilon_{co}-\varepsilon_{cm})}& ~~~,\varepsilon_{co}< \varepsilon\leqslant \varepsilon_{cm}
\end{matrix}\right.$$

El coeficiente $\gamma$ representa un porcentaje de la distancia al centroide del area bajo la curva con respecto al valor de la deformación $\varepsilon_c$

$$\gamma =\frac{\int_{0}^{\varepsilon_c}f_c \varepsilon d\varepsilon}{\varepsilon_c \int_{0}^{\varepsilon_c}f_c  d\varepsilon}$$

$$\gamma =\left\{\begin{matrix}
 \frac{8\varepsilon_{co}-3\varepsilon_c}{4(3\varepsilon_{co}-\varepsilon_c)} &, 0\leqslant \varepsilon \leqslant \varepsilon_{co} 
\\ 
 \frac{51{\varepsilon_{c}}^{2}\varepsilon_{co}-60{\varepsilon_{c}}^{2}\varepsilon_{cm}+6{\varepsilon_{c}}^{3}+10{\varepsilon_{co}}^{2}\varepsilon_{cm}-7{\varepsilon_{co}}^{3}}{\varepsilon_c(40\varepsilon_{co}\varepsilon_{cm}-31{\varepsilon_{co}}^{2}-120\varepsilon_{c}\varepsilon_{cm}+102\varepsilon_{c}\varepsilon_{co}+9{\varepsilon_{c}}^{2})}& ~~~,\varepsilon_{co}< \varepsilon\leqslant \varepsilon_{cm}
\end{matrix}\right.$$

Estos coeficientes serán de ayuda para evaluar la fuerza generada por los esfuerzos de compresión y la ubicación del punto de aplicación de la fuerza.

\\

Donde $f_s$ y $\varepsilon_s$ es el esfuerzo y la deformación unitaria en el acero, respectivamente, mientras que $\varepsilon _{y}$ es la deformación unitaria correspondiente a $f_y$ y $E_{s}$ es el módulo de elasticidad del acero.

# Determinación del Diagrama Momento Curvatura

Es posible deducir curvas teóricas momento-curvatura para secciones de concreto reforzado con flexión y carga axial, en base a suposiciones semejantes a las utilizadas para la determinaciónde la resistencia a flexión.

<div align="center"><img src=".\img\Esfuerzos.svg"
     width="600"></div>

La figura muestra una seccion de concreto armado reforzado. Para determinada deformación del concreto en la fibra extrema de compresión $\varepsilon_{c}$ y una profundidad $c$ del eje neutro, se pueden determinar las deformaciones del acero $\varepsilon_{s1},\varepsilon_{s2},\varepsilon_{s3},...,\varepsilon_{si}$ por triángulos semejantes del diagrama de deformaciones.


$$\varepsilon_{si}=\varepsilon_c\frac{(d_i-c)}{c}$$


<div align="center"><img src=".\img\Fuerzas.svg"
     width="300"></div>

Ahora se pueden encontrar los esfuerzos $f_{s1},f_{s2},f_{s3},...,f_{si}$ correspondientes a las deformaciones $\varepsilon_{s1},\varepsilon_{s2},\varepsilon_{s3},...,\varepsilon_{si}$ a partir de la curva esfuerzo-deformación del acero. En seguida se pueden encontrar las fuerzas del acero $S_{1},S_{2},S_{3},...,S_{i}$ a partir de los esfuerzos del acero y las áreas del mismo.

$$S_i=f_{si}A_{si}$$

Se puede encontrar la distribución del esfuerzo del concreto en la parte comprimida de la sección a partir del diagrama de deformaciones y la curva esfuerzo - deformación para el concreto. Para cualquier deformación dada del concreto $\varepsilon_{c}$ en la fibra extrema a compresión, se puede definir la fuerza de compresión del concreto $C$ y su posición en términos de los parámetros $\alpha$ y $\gamma$.

$$C=\alpha f'_cbc$$

En consecuencia, si se puede escribir el esfuerzo $f_c$ en el concreto en términos de la deformación $\varepsilon_c$, se puede determinar la fuerza del concreto y su línea de acción.

Se pueden escribir las ecuaciones de equilibrio como:

$$\sum S_i - C+P_a=0$$


$$M_r = C\gamma c+\sum S_i(d_i-c)+P_a\left ( \frac{h}{2}-c \right )$$

La curvatura es dada por:

$$\varphi =\frac{\varepsilon_c}{c}$$

Se puede determinar la relación teórica momento-curvatura para un nivel dado de carga axial, incrementando la deformación del concreto en la fibra $\varepsilon_c$ extrema a compresión. Para cada valor de $\varepsilon_c$ se encuentra la profundidad $c$ del eje neutro que satisface el equilibrio de las fuerzas ajustando $c$ hasta que las fuerzas internas calculadas cumplan dicha condición. Para nuestro caso de flexión solamente $P_a=0$, entonces se utilizan las fuerzas internas y la profundidad del eje neutro encontrados de esa manera para determinar el momento $M$ y curvatura $\varphi$ que correspondan a ese valor de $\varepsilon_c$.


# Código Elaborado

```python
# Unidades Base
m = 1
kg = 1
s = 1
# Otras Unidades
mm = 0.001*m
cm = 0.01*m
N = kg*m/s**2
kgf = 9.80665*N
tonf = 1000*kgf
ton= 1000*kg
Pa = N/m**2
MPa = 10**6*Pa
inch = 2.54*cm
ft = 12*inch
ksi = 6894757.2932*Pa
kip = ksi*inch**2
psi = 6894.76*Pa
# Constantes Físicas
g = 9.80665*m/s**2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Búsqueda de la solución del valor c mediante el algoritmo de la Bisección
def Buscar_solucion(Equilibrio,m,n,tol,εc,α,fc,b,εy,Es,As,fy,Pa):
    c = (m+n)/2
    EQ, εsi, fsi, Si, C,Si_di = Equilibrio(c,εc,α,fc,b,εy,Es,As,fy,Pa)
    EQm, εsim, fsim, Sim, Cm,Si_dim = Equilibrio(m,εc,α,fc,b,εy,Es,As,fy,Pa)

    while abs(EQ) > tol:
        if EQ*EQm < 0:
            n = c
        else:
            m = c
        c = (m+n)/2
    
        EQ, εsi, fsi, Si, C,Si_di = Equilibrio(c,εc,α,fc,b,εy,Es,As,fy,Pa)
        EQm, εsim, fsim, Sim, Cm,Si_dim = Equilibrio(m,εc,α,fc,b,εy,Es,As,fy,Pa)

    return c, εsi, fsi, Si, C,Si_di
```

```python
#Se plantea la condicion de equilibrio a evaluar
def Equilibrio(c,εc,α,fc,b,εy,Es,As,fy,Pa):

    εsi = np.zeros(len(As))
    fsi = np.zeros(len(As))
    Si = np.zeros(len(As))
    Si_di = np.zeros(len(As))

    for i in range(len(As)):
        εsi[i] = εc*(As[i,1]-c)/c

        if εsi[i] <= -εy:
            fsi[i] = -fy
        elif -εy< εsi[i] <=εy:
            fsi[i] = Es*εsi[i]
        else:
            fsi[i] = fy
        Si[i] = As[i,0]*fsi[i]
        Si_di[i] = Si[i]*(As[i,1]-c)

    C = α*fc*b*c
    EQ = sum(Si) + Pa - C

    return EQ, εsi, fsi, Si, C, Si_di
```

```python
#Se define la funcion Momento-Curvatura que inciara el cálculo de dichos valores
def Momento_curvatura(fc,εco,εcm,fy,Es,εy,εsu,b,h,Pa,As,Δc):

    intervalo = np.arange(Δc,εcm+Δc,Δc)
    φ = np.zeros(len(intervalo)+1)
    M = np.zeros(len(intervalo)+1)
    φ[0] = 0
    M[0] = 0

    for εc in intervalo:
        indice = np.where(intervalo == εc)
        ind = indice[0][0]
        if εc <= εco:
            α = εc*(3*εco-εc)/(3*εco**2)
            ϒ = (8*εco-3*εc)/(4*(3*εco-εc))
        else:
            α = (40*εco*εcm-31*εco**2-120*εc*εcm+102*εc*εco+9*εc**2)/(120*εc*(εco-εcm))
            ϒ = (51*(εc**2)*εco-60*(εc**2)*εcm+6*(εc**3)+10*(εco**2)*εcm-7*εco**3)/(εc*(40*εco*εcm-31*(εco**2)-120*εc*εcm+102*εc*εco+9*(εc**2)))
        
        c, εsi, fsi, Si, C,Si_di = Buscar_solucion(Equilibrio,Δc,h,0.0001,εc,α,fc,b,εy,Es,As,fy,Pa)
        
        Cϒc = C*ϒ*c
        F_Pa = Pa*(h/2-c)
        φ[ind+1] = εc/c
        M[ind+1] = (sum(Si_di) + F_Pa + Cϒc)
        
    return φ, M
```

```python
# PROPIEDADES DE LOS MATERIALES
fc = 320*kgf/cm**2
εco = 0.002
εcm = 0.004
fy = 4200*kgf/cm**2
Es = 2*10**6*kgf/cm**2
εy = 0.0021
εsu = 0.0756

# PROPIEDADES DE LA SECCIÓN
b = 23*cm
h = 46*cm
Pa = 0*ton  # Fuerza axial

# POSICIÓN DEL ACERO
#                  Asi  ,   di
As = np.array([[ 8*cm**2, 4*cm ],
               [ 4*cm**2, 16*cm ],
               [ 4*cm**2, 30*cm ],
               [ 8*cm**2, 42*cm ]])

# INTERVALO DE εc
Δc =0.0002

#LLAMADO A LA FUNCIÓN
φ1, M1 = Momento_curvatura(fc,εco,εcm,fy,Es,εy,εsu,b,h,Pa,As,Δc)
```

```python
#PLOTEAMOS LOS RESULTADOS
plt.figure(dpi=1080)
plt.plot(φ1,M1/tonf,label="Sección 1")
plt.xlabel(" φ(rad/m)")
plt.ylabel("M(ton.m)")
plt.title("Diagrama Momento - curvatura")
plt.legend()
plt.show()
```
